```python
# normal train-val-test split
import pandas as pd
from prophet import Prophet
import numpy as np

from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt

def compute_prophet_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    return {
        "MAE": mae,
        "RMSE": rmse,
    }

def time_split(df):
    dates = np.sort(df["ds"].unique())
    val_split_idx = int(len(dates) * 0.8)
    test_split_idx = int(len(dates) * 0.9)
    train_dates = dates[:val_split_idx]
    val_dates = dates[val_split_idx:test_split_idx]
    test_dates = dates[test_split_idx:]
    train_df = df[df["ds"].isin(train_dates)].copy()
    val_df = df[df["ds"].isin(val_dates)].copy()
    test_df = df[df["ds"].isin(test_dates)].copy()
    print(f"train={len(train_df)} - val={len(val_df)} - test={len(test_df)}")
    return train_df, val_df, test_df


files = [
    "/content/FPT_preprocessed.csv",
    "/content/HPG_preprocessed.csv",
    "/content/VCB_preprocessed.csv",
    "/content/VIC_preprocessed.csv",
    "/content/VNM_preprocessed.csv",
]

val_results = []
test_results = []

for file in files:
    df = pd.read_csv(file)
    code = df["stock_code"].iloc[0]

    window = 100
    df["rolling_std_close"] = df["close"].rolling(window).std()
    df["rolling_std_log_close"] = df["log_close"].rolling(window).std()

    plt.figure(figsize=(12,5))
    plt.plot(df["trading_date"], df["rolling_std_close"], label="Close")
    plt.plot(df["trading_date"], df["rolling_std_log_close"], label="Log Close")
    plt.title(f"{code} - {window}-Day Rolling Standard Deviation")
    plt.xlabel("Date")
    plt.ylabel("Rolling Standard Deviation")
    plt.legend()
    plt.grid(True)
    plt.show()

    for typee in ["log_close","close"]:
        data = df[["trading_date", typee]].copy()
        data.columns = ["ds", "y"]
        data["ds"] = pd.to_datetime(data["ds"])
        data = data.sort_values("ds").reset_index(drop=True)
        train_df, val_df, test_df = time_split(data)
        cps_list = []
        train_rmse_list = []
        val_rmse_list = []

        #tune cps trên val
        for cps in [0.01, 0.05, 0.1, 0.5, 1.0]:

            model = Prophet(
                daily_seasonality=False,
                weekly_seasonality=True,
                yearly_seasonality=True,
                changepoint_prior_scale=cps
            )

            model.fit(train_df)

            train_forecast = model.predict(train_df[["ds"]])
            train_rmse = np.sqrt(mean_squared_error(train_df["y"], train_forecast["yhat"]))

            val_forecast = model.predict(val_df[["ds"]])

            val_eval_df = val_df.copy().reset_index(drop=True)
            val_eval_df["yhat"] = val_forecast["yhat"].values
            val_eval_df["yhat_lower"] = val_forecast["yhat_lower"].values
            val_eval_df["yhat_upper"] = val_forecast["yhat_upper"].values

            val_metrics = compute_prophet_metrics(
                val_eval_df["y"].values,
                val_eval_df["yhat"].values
            )

            cps_list.append(cps)
            train_rmse_list.append(train_rmse)
            val_rmse_list.append(val_metrics["RMSE"])

            val_results.append({
                "Model": "Prophet",
                "Ticker": code,
                "Target": typee,
                "CPS": cps,
                "Train_Size": len(train_df),
                "Val_Size": len(val_df),
                "Train_RMSE": train_rmse,
                "Val_MAE": val_metrics["MAE"],
                "Val_RMSE": val_metrics["RMSE"],
            })

            print(f"Validation - Prophet - {code} - {typee} - cps={cps}")
            print(f"Train: {train_df['ds'].min().date()} -> {train_df['ds'].max().date()}")
            print(f"Val  : {val_eval_df['ds'].min().date()} -> {val_eval_df['ds'].max().date()}")
            print(f"Val MAE : {val_metrics['MAE']:.6f}")
            print(f"Val RMSE: {val_metrics['RMSE']:.6f}")
        plt.figure(figsize=(8, 5))
        plt.plot(cps_list, train_rmse_list, marker="o", label="Train RMSE")
        plt.plot(cps_list, val_rmse_list, marker="s", label="Validation RMSE")
        plt.xscale("log")
        plt.xlabel("changepoint_prior_scale")
        plt.ylabel("RMSE")
        plt.title(f"{code} - {typee}: CPS tuning on validation set")
        plt.legend()
        plt.grid(True)
        plt.show()

        best_idx = np.argmin(val_rmse_list)
        best_cps = cps_list[best_idx]

        print(f"Best CPS for {code}: {best_cps}")

        #train lại trên cả train và val
        final_train_df = pd.concat([train_df, val_df], axis=0).sort_values("ds").reset_index(drop=True)

        final_model = Prophet(
            daily_seasonality=False,
            weekly_seasonality=True,
            yearly_seasonality=True,
            changepoint_prior_scale=best_cps
        )

        final_model.fit(final_train_df)

        test_forecast = final_model.predict(test_df[["ds"]])

        test_eval_df = test_df.copy().reset_index(drop=True)
        test_eval_df["yhat"] = test_forecast["yhat"].values
        test_eval_df["yhat_lower"] = test_forecast["yhat_lower"].values
        test_eval_df["yhat_upper"] = test_forecast["yhat_upper"].values

        test_metrics = compute_prophet_metrics(
            test_eval_df["y"].values,
            test_eval_df["yhat"].values
        )

        test_results.append({
            "Model": "Prophet",
            "Ticker": code,
            "Target": typee,
            "Best_CPS": best_cps,
            "Final_Train_Size": len(final_train_df),
            "Test_Size": len(test_eval_df),
            "Test_MAE": test_metrics["MAE"],
            "Test_RMSE": test_metrics["RMSE"],
        })

        path = f"/content/prophet_final_test_{code}_{typee}_bestcps{best_cps}.csv"
        test_eval_df.to_csv(path, index=False)

        plt.figure(figsize=(12, 5))
        plt.plot(test_eval_df["ds"], test_eval_df["y"], label="Actual log_close")
        plt.plot(test_eval_df["ds"], test_eval_df["yhat"], label="Prophet Predicted log_close")
        plt.fill_between(
            test_eval_df["ds"],
            test_eval_df["yhat_lower"],
            test_eval_df["yhat_upper"],
            alpha=0.2,
            label="Confidence interval"
        )
        plt.title(f"Final Test Forecast vs Actual - {code} - best CPS={best_cps}")
        plt.xlabel("Date")
        plt.ylabel("log_close")
        plt.legend()
        plt.grid(True)
        plt.show()

        print(f"FINAL TEST - Prophet - {code} - {typee}")
        print(f"Final Train: {final_train_df['ds'].min().date()} -> {final_train_df['ds'].max().date()}")
        print(f"Test       : {test_eval_df['ds'].min().date()} -> {test_eval_df['ds'].max().date()}")
        print(f"Best CPS   : {best_cps}")
        print(f"Test MAE   : {test_metrics['MAE']:.6f}")
        print(f"Test RMSE  : {test_metrics['RMSE']:.6f}")


val_results_df = pd.DataFrame(val_results)
test_results_df = pd.DataFrame(test_results)

val_results_df.to_csv("/content/prophet_validation_results.csv", index=False)
test_results_df.to_csv("/content/prophet_final_test_results.csv", index=False)

print("\nValidation results:")
print(val_results_df)

print("\nFinal test results:")
print(test_results_df)
```

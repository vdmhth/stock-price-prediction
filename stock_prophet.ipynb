```python
import pandas as pd
from prophet import Prophet

files = [
    "/content/FPT_preprocessed.csv",
    "/content/HPG_preprocessed.csv",
    "/content/VCB_preprocessed.csv",
    "/content/VIC_preprocessed.csv",
    "/content/VNM_preprocessed.csv",
]

for file in files:
  df = pd.read_csv(file)
  code = df["stock_code"].iloc[0]
  for typee in ["close","log_close"]:
    for range in [1,5,20]:
      data = df[["trading_date", f"{typee}"]].copy()

      data.columns = ["ds", "y"]
      data["ds"] = pd.to_datetime(data["ds"])

      model = Prophet(
          daily_seasonality=False,
          weekly_seasonality=True,
          yearly_seasonality=True,
          changepoint_prior_scale=1.0
      )

      model.fit(data)

      future = model.make_future_dataframe(periods=range, freq="B")
      forecast = model.predict(future)

      check = pd.DataFrame({
          "ds": forecast["ds"],
          "y": data["y"],
          "yhat": forecast["yhat"],
          "yhat_lower": forecast["yhat_lower"],
          "yhat_upper":forecast["yhat_upper"]
      })

      path=f"/content/prophet_{range}d_{code}_{typee}.csv"
      check.to_csv(path,index=False)
      print(check.tail(20))
```

```python
def format_pvalue(p):
    if pd.isna(p):
        return ""
    if p < 0.001:
        return "<0.001"
    return f"{p:.4f}"


df_report = df_autocorr[report_cols].copy()

df_report["ARCH effect"] = np.where(
    df_report["arch_5_effect"] & df_report["arch_20_effect"],
    "Yes",
    "Mixed/No"
)

df_report = df_report.drop(columns=["arch_5_effect", "arch_20_effect"])

pval_cols = [
    "lb_5_pval",
    "lb_20_pval",
    "lb_abs_5_pval",
    "lb_abs_20_pval",
    "arch_5_pval",
    "arch_20_pval",
]

for col in pval_cols:
    df_report[col] = df_report[col].apply(format_pvalue)

df_report = df_report.rename(columns={
    "series": "Series",
    "lb_5_pval": "LB return p(5)",
    "lb_20_pval": "LB return p(20)",
    "lb_abs_5_pval": "LB |return| p(5)",
    "lb_abs_20_pval": "LB |return| p(20)",
    "arch_5_pval": "ARCH-LM p(5)",
    "arch_20_pval": "ARCH-LM p(20)",
})

display(df_report)
```
